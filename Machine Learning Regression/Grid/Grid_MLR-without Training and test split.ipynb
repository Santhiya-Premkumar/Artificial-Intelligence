```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
# without training and test Split
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
independent = sc.fit_transform(independent)
#  (Xi-Xmean)/Xstd
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
param_grid= {'fit_intercept':[True,False],'copy_X':[True,False],'n_jobs':[None],'positive':[True,False]}
grid = GridSearchCV(LinearRegression(), param_grid, refit = True, verbose = 3,n_jobs=-1)
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_
re
```

```python
grid_predictions=grid.predict(independent)
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(dependent,grid_predictions)
print(r_score)
print("The R_score value for best parameter {}:".format(grid.best_params_))
```
